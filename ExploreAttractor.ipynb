# Runge-Kutta and Euler Methods 
   # An experiement to answer the following questions 
    1. How does the plotted solution depend on your choice of time step size, and
    your choice of increment? Do you see an improvement in precision from
    using the higher-order integration methods?
    2. How does the solution depend upon the initial conditions [x0, y0, z0]? How
    small a change can you make while still reproducing roughly the same
    dynamical curves?
    

    # For the first example, the program is run with the following scenario :-
        A) We run the program for Euler & order 4 methods. The time range is 0 to 100 and the step size is 100 & 10000.
            a) Euler method, step size 100
            b) Order 4 method, step size 100
            c) Euler method, step size 10000
            d) Order 4 method, step size 10000
        B) We run the program for Euler & order 4 methods. The time range is 0 to 1000 and the step size is 100 & 10000.
            a) Euler method, step size 100
            b) Order 4 method, step size 100
            c) Euler method, step size 10000
            d) Order 4 method, step size 10000
     # Based on the analysis we find there is higher precision when leveraging higher order integration methods.

```python
from attractor import Attractor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

a = Attractor()
a.points=100
a.start=0
a.end=100
a.evolve(r0=[10,28,8/3],order=1)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()

b = Attractor()
b.points=100
b.start=0
b.end=100
b.evolve(r0=[10,28,8/3],order=2)
b.plotx()
b.ploty()
b.plotz()
b.plotxy()
b.plotyz()
b.plotzx()
b.plot3d()

c = Attractor()
c.points=100
c.start=0
c.end=100
c.evolve(r0=[10,28,8/3],order=4)
c.plotx()
c.ploty()
c.plotz()
c.plotxy()
c.plotyz()
c.plotzx()
c.plot3d()
```

```python
a = Attractor()
a.points=10000
a.start=0
a.end=100
a.evolve(r0=[10,28,8/3],order=1)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()

b = Attractor()
b.points=10000
b.start=0
b.end=100
b.evolve(r0=[10,28,8/3],order=2)
b.plotx()
b.ploty()
b.plotz()
b.plotxy()
b.plotyz()
b.plotzx()
b.plot3d()

c = Attractor()
c.points=10000
c.start=0
c.end=100
c.evolve(r0=[10,28,8/3],order=4)
c.plotx()
c.ploty()
c.plotz()
c.plotxy()
c.plotyz()
c.plotzx()
c.plot3d()
```
